# QB Career Stats Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import math
```

```python
df = pd.read_csv('data/raw/Career_Stats_Passing.csv')
df.head()
```

```python
df.info()
```

```python
# replace '--'
df.replace({'--': 0},inplace=True)
df.head()
```

```python
# rename cols
def norm_cols(cols):
    '''
    converts column naming to traditional database-stle naming convention
    '''
    
    cols = cols.str.replace(' ','_').str.replace('(','').str.replace(')','').str.lower()
    return cols
```

```python
# apply column renaming to dataframe
df.columns = norm_cols(df.columns)
df.columns
```

```python
df['passing_yards'] = df['passing_yards'].str.replace(',','')
```

```python
df['passing_yards'] = pd.to_numeric(df['passing_yards'])
```

```python
# most passing yards
df.groupby(['player_id','name'])[['games_played','passing_yards']].sum().sort_values('passing_yards',ascending=False)[0:11]
```

```python
top_qbs = df.groupby(['player_id','name'])[['games_played','passing_yards']].sum().sort_values('passing_yards',ascending=False)[0:11]
top_qbs
```

```python
top_qbs['yards_per_game'] = round(top_qbs['passing_yards'] / top_qbs['games_played'], 2)
top_qbs
```

```python
top_qbs['trailing_leader_by'] = top_qbs['passing_yards'][0] - top_qbs['passing_yards']
top_qbs
```
